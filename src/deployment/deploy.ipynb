```python
import cv2
from tensorflow.keras.models import load_model
import numpy as np

# Load model dari file .h5
model_path = 'model_1.h5'
model = load_model(model_path)

# Label kelas
class_labels = ['mengantuk & menguap',
 'mengantuk & tidak menguap',
 'menguap & tidak mengantuk']

# Fungsi untuk memproses frame dan melakukan prediksi
def process_frame(frame):
    # Resize frame ke ukuran input model
    resized_frame = cv2.resize(frame, (64, 64))
    # Normalize pixel values
    normalized_frame = resized_frame / 255.0
    # Tambahkan batch dimension
    input_frame = np.expand_dims(normalized_frame, axis=0)
    
    # Prediksi kelas
    prediction = model.predict(input_frame)
    predicted_class = np.argmax(prediction, axis=1)[0]
    
    return class_labels[predicted_class]

# Inisialisasi webcam
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Tidak dapat mengakses kamera.")
else:
    print("Mengakses kamera...")

while True:
    # Baca frame dari webcam
    ret, frame = cap.read()
    
    if not ret:
        print("Error: Tidak dapat membaca frame dari kamera.")
        break
    
    # Lakukan prediksi pada frame
    predicted_class = process_frame(frame)
    
    # Tampilkan hasil prediksi pada frame
    cv2.putText(frame, f'Prediksi: {predicted_class}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
    
    # Tampilkan frame
    cv2.imshow('Live Video', frame)
    
    # Hentikan loop jika tombol 'q' ditekan
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video capture dan tutup semua jendela OpenCV
cap.release()
cv2.destroyAllWindows()
```
